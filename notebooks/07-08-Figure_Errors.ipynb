## 1. Fullspace with skindepth

Here, we compare the digital filter with the analytical fullspace solution for a wide range of frequencies and resistivities (simple case: $\lambda = 1$, $\varepsilon_r = 10$, $\mu_r=1$). 

We define the offset as a function of the skin depth $z_{skin} = 1/\beta$, defined by (e.g., Ward and Hohmann, 1988, equation 1.49)

$$
  z_{skin} = \left\{\frac{\omega^2 \varepsilon\mu}{2}
    \left[ \sqrt{1 + \frac{1}{\left(\omega\varepsilon\rho\right)^2}} -1\right]
  \right\}^{-1/2} \, .
$$

**Ward, S. H., and G. W. Hohmann, 1988**, Electromagnetic theory for geophysical applications: SEG, Investigations in Geophysics, No. 3, 4, 130-131. [doi: 10.1190/1.9781560802631](https://doi.org/10.1190/1.9781560802631).

```python
# Define powers of frequency
fexp = np.arange(-10., 14.1, 0.25)

# Define powers of resistivity
rexp = np.arange(-10, 10.1, 0.25)

# Electric permittivity
eperm_r = 10               # Relative electric permittivity
eperm = epsilon_0*eperm_r  # Total electri permittivity

# Fixed parameters
params = {'src': [0, 0, 0],
          'depth': [],
          'epermH': eperm_r,
          'epermV': eperm_r,
          'ht': 'fht',
          'htarg': 'wer_201_2018',
          'verb': 1}

def calc_data():

    # Pre-allocate output arrays
    error = np.ones((len(fexp), len(rexp)))
    ansol = np.ones((len(fexp), len(rexp)))
    offs = np.ones((len(fexp), len(rexp)))
    
    # Loop over frequencies
    for i, val in enumerate(fexp):

        # Get frequency
        freq = 10**val

        # Loop over resistivities
        for ii, val2 in enumerate(rexp):

            # Get this resistivity
            res = 10**val2

            # Calculate skin depth and set to offset
            omega = 2*np.pi*freq
            off = np.sqrt(2/(mu_0*eperm*omega**2))/np.sqrt(np.sqrt(1 + (1/(res*omega*eperm))**2) - 1)
            offs[i, ii] = off

            # Define this receiver
            rec = [off, 0, off/10]

            # Get analytical result (xdirect=True) and store
            an_out = epm.dipole(rec=rec, res=res, freqtime=freq, xdirect=True, **params)
            ansol[i, ii] = np.abs(an_out)

            # Get filter result (xdirect=False), store relative error
            dlf_out = epm.dipole(rec=rec, res=res, freqtime=freq, xdirect=False, **params)
            error[i, ii] = np.abs((dlf_out-an_out)/an_out)

    return error, ansol, offs
```

```python
def plot_data(error, ansol, offs):
    # Figure
    plt.figure(figsize=(7, 4))

    # Get x- and y- axis
    x = np.r_[rexp-0.125, rexp[-1]+0.125]
    y = np.r_[fexp-0.125, fexp[-1]+0.125]
    extent = [x[0], x[-1], y[0], y[-1]]
    
    # Calculate omega*epsilon*rho = 1
    oer1 = np.log10(1/(2*np.pi*eperm*10**x))
    
    # Plot amplitude
    ax1 = plt.subplot(121)
    plt.title('(a) Analytical solution')

    CP1 = plt.pcolormesh(x, y, np.log10(ansol), linewidth=0, rasterized=True)

    CS1 = plt.contour(np.log10(offs), colors='0.7', linestyles='solid', extent=extent)
    plt.clabel(CS1, fmt='%1i', inline_spacing=25)
    
    # plot omega*epsilon*rho = 1
    plt.plot(x, oer1, 'C3', lw=2)
    
    plt.axis(extent)
    plt.xlabel(r'$\log_{10}$ resistivity $(\Omega\,$m)')
    plt.ylabel(r'$\log_{10}$ frequency (Hz)')
    plt.yticks(fexp[::16])
    plt.xticks(rexp[::16])

    # Colorbar
    cax, kw = plt.matplotlib.colorbar.make_axes(ax1, location='bottom', pad=0.2)
    cb = plt.colorbar(CP1, cax=cax, **kw)
    cb.set_label(r'$\log_{10}$ amplitude (V/m)')

    # Plot error
    ax2 = plt.subplot(122)
    plt.title('(b) Error')

    CP2 = plt.pcolormesh(x, y, np.log10(error), linewidth=0, rasterized=True)

    CS2 = plt.contour(np.log10(offs), colors='0.7', linestyles='solid', extent=extent)
    plt.clabel(CS2, fmt='%1i', inline_spacing=25)

    # plot omega*epsilon*rho = 1
    plt.plot(x, oer1, 'C3', lw=2)
    
    plt.axis(extent)
    ax2.yaxis.tick_right()
    ax2.yaxis.set_label_position("right")
    plt.xlabel(r'$\log_{10}$ resistivity $(\Omega\,$m)')
    plt.ylabel(r'$\log_{10}$ frequency (Hz)')
    plt.yticks(fexp[::16])
    plt.xticks(rexp[::16])

    # Colorbar
    cax2, kw2 = plt.matplotlib.colorbar.make_axes(ax2, location='bottom', pad=0.2)
    cb2 = plt.colorbar(CP1, cax=cax2, **kw2)
    cb2.set_label('$\log_{10}$ relative error ($-$)')

    # Save figure
    if save:
        for ending in ['png', 'pdf', 'eps']:
            plt.savefig('../figures/Figure-8.'+ending, bbox_inches='tight')
    
    # Title
    #title = 'Horizontal offset = $z_{skin}$; vertical src-rec separation = $z_{skin}/10$.'
    #plt.suptitle(title, y=.98)

    plt.show()
```

```python
error, ansol, offs = calc_data()
```

```python
plot_data(error, ansol, offs)
```

## 2. Deep water, shallow water, and land case

### Generate data

```python
# Offsets from 50 m to 20 km, logarithmically
x = np.logspace(np.log10(50/200), 2, 50)*200

# Frequencies from 0.01 Hz to 10 Hz, logarithmically
freqs = np.logspace(-2, 1, 50)

# Resistivities of lower half-space
res = [1, 10, 100, 1000]

# Source and receiver depths for the three cases
zsrc = [10, 1990, 0.5]
zrec = [100, 2000, 0.8]

# Water depths and resistivity for deep/shallow cases
wdepth = [400, 2000]
wres = 0.3

# General input parameters
inp = {'ab': 11, 'freqtime': freqs, 'verb': 1,}  

# Pre-allocate arrays for all three models, 4 resistivities
amp = np.zeros((3, len(freqs), len(x), 4))
err = np.zeros((3, len(freqs), len(x), 4))

# Loop over the three cases
for i, name in enumerate(['deep', 'shallow', 'land']):
    
    # Update model
    inp['src'] = [0, 0, zrec[i]]
    inp['rec'] = [x, x*0, zsrc[i]]
    if i < 2:
        inp['depth'] = [0, wdepth[i]]
        inp['res'] = [2e14, wres, -1]  # -1: we'll update later
        inp['epermH'] = [0, 1, 1]
        inp['epermV'] = [0, 1, 1]
    else:
        inp['depth'] = [0]
        inp['res'] = [2e14, -1]  # -1: we'll update later
        inp['epermH'] = [0, 1]
        inp['epermV'] = [0, 1]

    # Loop over 4 lower halfspace resistivities
    for ii, ires in enumerate(res):

        # Update lower halfspace resistivity
        inp['res'][-1] = ires

        # Calculate responses
        wer201 = epm.dipole(htarg='wer_201_2018', **inp)
        qwe = epm.dipole(ht='qwe', **inp)

        # Calculate relative error
        error = np.abs((wer201-qwe)/qwe)

        # Store amplitude and relative error
        amp[i, :, :, ii] = np.log10(np.abs(qwe))
        err[i, :, :, ii] = np.log10(error)
```

### Plot relative error

```python
# pcolor needs start and endpoints, not actual points (midpoints)
xdif = (np.log10(x[1])-np.log10(x[0]))/2
xax = np.r_[np.log10(x/1000)-xdif, np.log10(x[-1]/1000)+xdif]
fdif = (np.log10(freqs[1])-np.log10(freqs[0]))/2
fax = np.r_[np.log10(freqs)-fdif, np.log10(freqs[-1])+fdif]
```

```python
# Get min and max    
vmin = np.min(err)  # Take overall lowest value as minimum
vmax = 0            # Limit this to 100 % relative error

# General parameters
params = {'vmin': vmin, 'vmax': vmax, 'linewidth': 0, 'rasterized': True}

# Start figure
fig, axs = plt.subplots(figsize=(7.5, 4.3), nrows=3, ncols=4)
plt.subplots_adjust(wspace=0.05, hspace=0.07)
axs = axs.ravel()
cplist = []

# Loop over cases
for i, name in enumerate(['2 km water', '400 m water', 'Land']):

    # Loop over resistivities
    for ii, ires in enumerate(res):
        plt.sca(axs[i*4+ii])
        
        # Plot result, store handle
        cp = plt.pcolormesh(xax, fax, err[i, :, :, ii], **params)
        cplist.append(cp)
            
        extent = [xax[0], xax[-1], fax[0], fax[-1]]
        CS2 = plt.contour(amp[i, :, :, ii], [-24, -21, -18, -15, -12, -9, -6, -3],
                          colors='0.7', linestyles='solid', extent=extent)
        plt.clabel(CS2, fmt='%1i', inline_spacing=25)

        # title
        if i == 0:
            plt.title(r'$\rho =$ '+str(ires)+r'$\,\Omega\,$m', fontsize=8, y=1.05)
                
        # x-ticks and labels
        if i == 2:
            plt.xticks([-1, 0, 1], ['0.1', '1', '10'])
        else:
            plt.xticks([-1, 0, 1], ())

        # y-label
        axs[i*4+ii].yaxis.tick_right()

        if ii == 3:
            plt.yticks([-2, -1, 0, 1], ['0.01', '0.1', '1', '10'])
            if i == 1:
                axs[i*4+ii].yaxis.set_label_position("right")
                plt.ylabel(r'Frequency (Hz)', labelpad=10)
        elif ii == 0:
            plt.ylabel(name, labelpad=15)
            plt.yticks([-2, -1, 0, 1], ())
        else:
            plt.yticks([-2, -1, 0, 1], ())

        # Show ticks on all four sides
        axs[i*4+ii].xaxis.set_ticks_position('both')
        axs[i*4+ii].yaxis.set_ticks_position('both')
            
# xlabel
fig.text(0.47, 0.02, 'Offset (km)', ha='center')
            
# Colorbar
cax, kw = plt.matplotlib.colorbar.make_axes(axs,# location='bottom',
                                            fraction=.02, pad=0.15, aspect=20)
cb = plt.colorbar(cplist[(i+1)*4-1], cax=cax, **kw)
cb.set_label(r'$\log_{10}$ relative error ($-$)')#, labelpad=-40, y=1.05, rotation=0)

# Save figure
if save:
    for ending in ['png', 'pdf', 'eps']:
        plt.savefig('../figures/Figure-7.'+ending, bbox_inches='tight')

# Title
# plt.suptitle('Relative error ($-$)', fontsize=14, y=0.97)

plt.show()
```
